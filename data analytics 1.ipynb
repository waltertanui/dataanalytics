```python
import pandas as pd
```

```python
all_data = pd.read_csv(r"F:\python\marathon data\all_data.csv")
all_data.head()
```

```python
###clean up the data
#drop NAN
rows_with_nan = all_data[all_data.isna().any(axis=1)]
print(rows_with_nan)
```

## drop all NAN

```python
## drop all NAN
all_data_clean = all_data.dropna()
```

## chech if all NAN is droped

```python
## chech if all NAN is droped
rows_with_nan = all_data_clean[all_data_clean.isna().any(axis=1)]
rows_with_nan.head()
```

```python
all_data_clean.head()
```

Add a city column

```python
all_data_clean= all_data_clean[all_data_clean['Order Date'].str[0:2] != 'Or']
all_data_clean.head()
```

## convert colums to the correct type

```python
# Convert 'Quantity Ordered' column to integers
all_data_clean['Quantity Ordered'] = all_data_clean['Quantity Ordered'].astype(int)

# Convert 'Price Each' column to floats
all_data_clean['Price Each'] = all_data_clean['Price Each'].astype(float)
```

```python
# add month column
all_data_clean['Month'] = all_data_clean['Order Date'].str[0:2]
all_data_clean['Month'] = all_data_clean['Month'].astype('int32')
all_data_clean.head()
```

###add a sales column

```python
all_data_clean['Sales'] = all_data_clean['Quantity Ordered'] * all_data_clean['Price Each']
all_data_clean.head()
```

add a city column

```python
all_data_clean['City'] = all_data_clean['Purchase Address'].apply(lambda x: x.split(',')[1])
all_data_clean['State'] = all_data_clean['Purchase Address'].apply(lambda x: x.split(',')[2].split(' ')[1])
all_data_clean['City'] = all_data_clean['City'] + ', ' + all_data_clean['State']
all_data_clean.drop(columns=['State'], inplace=True)

all_data_clean.head()
```

q1: what was the best month for sales? how much was earned that month?

```python
results = all_data_clean.groupby('Month').sum()
results.head()
```

```python
import matplotlib.pyplot as plt

months = range(1, 13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('sales in USD ($)')
plt.xlabel('Mounts')
plt.show()
```

# what city has higest number of sales

```python
results = all_data_clean.groupby('City').sum()
results.head()
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in all_data_clean.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation = 'vertical', size = 8)
plt.ylabel('sales in USD ($)')
plt.xlabel('cities')
plt.show()
```

q3 what time should we diplay our products to maximize our reach

```python
all_data_clean.head()
```

```python
all_data_clean['Order Date'] = pd.to_datetime(all_data_clean['Order Date'])
all_data_clean['Hour'] = all_data_clean['Order Date'].dt.hour
all_data_clean['Minute'] = all_data_clean['Order Date'].dt.minute
all_data_clean.head()
```

```python
import matplotlib.pyplot as plt


hours = [hour for hour, df in all_data_clean.groupby('Hour')]

plt.plot(hours, all_data_clean.groupby(['Hour']).count())
plt.xlabel('Hours')
plt.ylabel('Number of order')
plt.xticks(hours)
plt.grid()
plt.show()
```
